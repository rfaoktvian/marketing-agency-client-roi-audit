### IMPORT LIBRARY

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

### LOAD DATA

```python
DATA_DIR = Path("../data/raw")

def load_table(filename: str) -> pd.DataFrame:
    df = pd.read_csv(DATA_DIR / filename)
    print(f"{filename}: {df.shape[0]} rows, {df.shape[1]} cols")
    return df

clients     = load_table("clients.csv")
campaigns   = load_table("campaigns.csv")
ad_spend    = load_table("ad_spend.csv")
leads       = load_table("leads.csv")
conversions = load_table("conversions.csv")
```
